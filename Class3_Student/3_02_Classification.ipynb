```python
import requests

url="https://raw.githubusercontent.com/capigian/BIA6920/master/Class3_Student/data/affairsdata.csv"
affairs=pd.read_csv(url, index_col = 0)
```

```python
affairs['age_num'] = affairs.Age.str.split('\s+').str[0]
```

```python
affairs.info()
```

```python
affairs['Years_Married'] = affairs.Years_Married.str.split('\s+').str[0]
```

```python
affairs.tail(25)
```

```python
affairs['Years_Married'].astype(str).astype(float)
```

```python
def child(c):
  if c['Children?'] == "yes":
    return 1
  else:
    return 0

affairs['Children?'] = affairs.apply(child, axis=1)
```

```python
affairs.info(25)
```

```python
def affair(c):
  if c['Number of Affairs in Past Year'] > 0:
    return 1
  else:
    return 0

affairs['Number of Affairs in Past Year'] = affairs.apply(affair, axis=1)
```

```python
# "Sex" Coulumn has male/feamle as value. We can use LabelEncoder
# to convert these to int. male:1,female:0
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
affairs['Gender'] = lb.fit_transform(affairs['Gender']) 
```

```python
affairs.head(25)
```

```python
def marriage(c):
    if c['Marriage_Rating'] == "Very unhappy":
        return 0
    elif c['Marriage_Rating'] == "Somewhat unhappy":
        return 1
    elif c['Marriage_Rating'] == "Average":
        return 2
    elif c['Marriage_Rating'] == "Happier than average":
        return 3
    else:
        return 4
        
        
affairs['Marriage_Rating'] = affairs.apply(marriage, axis=1)
```

```python
affairs.head()
```

```python
affairs['Occupation'] = lb.fit_transform(affairs['Occupation']) 
```

```python
affairs.head()
```

```python
religious = pd.read_csv('data/religious.csv', header=0)
affairs = pd.merge(affairs, religious, how='left', on='Religious')
```

```python
affairs.info()
```

```python
affairs.describe()
```

```python
affairs.head(15)
```

```python
education = pd.read_csv('data/education.csv', header=0)
affairs = pd.merge(affairs, education, how='left', on='Education')
```

```python
affairs.head()
```

```python
affairs = affairs.drop(['Age','Religious', 'Education'], axis=1) 
```

```python
affairs=affairs.rename(columns = {'age_num':'Age', 'religious_num':'Religion', 'education_num':'Education', 'Number of Affairs in Past Year':'AffairsYear'})
```

```python
affairs.head()
```

```python
affairs = affairs[['Gender', 'Years_Married', 'Children?', 'Occupation', 'Marriage_Rating', 'Age', 'Religion', 'Education', 'AffairsYear']]
```

```python
affairs.head()
```

```python
Y = affairs['AffairsYear']
X = affairs = affairs[['Gender', 'Years_Married', 'Children?', 'Marriage_Rating', 'Age', 'Religion', 'Education']]
```

```python
#Generating decision tree in a file(image)
from sklearn import tree
from io import StringIO
from IPython.display import Image, display
```
